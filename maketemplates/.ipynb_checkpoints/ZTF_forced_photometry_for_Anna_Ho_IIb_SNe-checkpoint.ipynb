```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'data/ZTF_ForcedPhot/'
```

```python
peaks = {'SN2020xlt':59141.45,
         'SN2018gjx':58379.44,
         'SN2020rsc':59081.47,
         'SN2019rta':58759.43,
         'SN2020ano':58871.45,
         'SN2020ikq':58971.30
        }
```

```python
sn_name = 'SN2020xlt'
filepath = path+'{}_lc.txt'.format(sn_name)
```

```python
# Using readlines()
file1 = open(filepath, 'r')
Lines = file1.readlines()
```

```python
for line in Lines:
    if line.startswith('#'):
        continue
    if line.startswith(' index'):
        columns = line.split()
#     print(line)
#     else:
#         df.loc[len(df.index)] = line.split()
```

```python
for i,c in enumerate(columns):
    columns[i] = c.strip().split(',')[0]
```

```python
df = pd.DataFrame(columns=columns)
```

```python
for line in Lines:
    if line.startswith('#'):
        continue
    if line.startswith(' index'):
        continue
    df.loc[len(df.index)] = line.split()
```

```python
df['forcediffimflux'][df['forcediffimflux']=='null']= np.nan
df['forcediffimfluxunc'][df['forcediffimfluxunc']=='null']= np.nan
df['forcediffimchisq'][df['forcediffimchisq']=='null']= np.nan
df['forcediffimsnr'][df['forcediffimsnr']=='null']= np.nan
df['nearestrefmag'][df['nearestrefmag'] == 'null']= np.nan 
df['nearestrefmagunc'][df['nearestrefmagunc'] == 'null'] = np.nan
```

```python
df['jd'] = df['jd'].astype(float)
df['forcediffimflux'] = df['forcediffimflux'].astype(float)
df['forcediffimfluxunc'] = df['forcediffimfluxunc'].astype(float)
df['forcediffimchisq'] = df['forcediffimchisq'].astype(float)
df['forcediffimsnr'] = df['forcediffimsnr'].astype(float)
df['zpdiff'] = df['zpdiff'].astype(float)
df['nearestrefmag'] = df['nearestrefmag'].astype(float)
df['nearestrefmagunc'] = df['nearestrefmagunc'].astype(float)
df['scisigpix'] = df['scisigpix'].astype(float)
```

```python
peak = peaks[sn_name]
```

```python
SNT = 0
plt.scatter(df['jd'][(df['filter']=='ZTF_g')]-2400000.5-peak, 
             df['forcediffimflux'][(df['filter']=='ZTF_g')],
            marker='.',
            color='green')
plt.scatter(df['jd'][(df['filter']=='ZTF_r')]-2400000.5-peak, 
             df['forcediffimflux'][(df['filter']=='ZTF_r')],
            marker='.',
            color='red')
plt.scatter(df['jd'][(df['filter']=='ZTF_i')]-2400000.5-peak, 
             df['forcediffimflux'][(df['filter']=='ZTF_i')],
            marker='.',
            color='purple')

# peak = 58871.45
# plt.xlim(peak-25, peak+75)
plt.xlabel('Time (MJD)')
plt.ylabel('Magnitude')
plt.title(sn_name)
plt.axvline(0)
plt.axhline(0)

# plt.gca().invert_yaxis()
# plt.savefig(path+sn_name+'_SNR.png')
```

```python
df['nearestrefflux'] = 10**(0.4 * (df['zpdiff'] - df['nearestrefmag']))
df['nearestreffluxunc'] = df['nearestrefmagunc'] * df['nearestrefflux'] / 1.0857
```

```python
df['Fluxtot'] = df['forcediffimflux'] + df['nearestrefflux']
df['Fluxunctot'] = np.sqrt( df['forcediffimfluxunc']**2 - df['nearestreffluxunc']**2)
df['SNRtot'] = df['Fluxtot'] / df['Fluxunctot']
```

```python
SNT = 3
SNU = 5
```

```python
df['mag2'] = np.zeros(len(df))
df['mag2unc'] = np.zeros(len(df))

for i, f in enumerate(df['Fluxtot'].values):
    if df.iloc[i]['SNRtot']>SNT:
#         print('yes')
        
        df['mag2'].loc[i] = df.iloc[i]['zpdiff'] -\
                                            2.5*np.log10(f)
        df['mag2unc'].loc[i] = 1.0857 / df.iloc[i]['SNRtot']
    else:
        df['mag2'].loc[i] = df.iloc[i]['zpdiff'] -\
        2.5*np.log10(SNU*f)
```

```python
df['mag'] = df['zpdiff'] - 2.5*np.log10(df['Fluxtot'])
df['magunc'] = 1.0857 / df['SNRtot']
```

```python
df['magunc'][df['magunc']<0]= np.nan
```

```python
fig, axs = plt.subplots(2,1, figsize=(5, 10))

axs[0].errorbar(df['jd'][(df['filter']=='ZTF_g')&(df['SNRtot']>3)]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_g')&(df['SNRtot']>3)],
             yerr = df['magunc'][(df['filter']=='ZTF_g')&(df['SNRtot']>3)],
            marker='.',
            color='green')
axs[0].errorbar(df['jd'][(df['filter']=='ZTF_r')&(df['SNRtot']>3)]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_r')&(df['SNRtot']>3)],
             yerr = df['magunc'][(df['filter']=='ZTF_r')&(df['SNRtot']>3)],
            marker='.',
            color='red')
axs[0].errorbar(df['jd'][(df['filter']=='ZTF_i')&(df['SNRtot']>3)]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_i')&(df['SNRtot']>3)],
             yerr = df['magunc'][(df['filter']=='ZTF_i')&(df['SNRtot']>3)],
            marker='.',
            color='purple')


axs[1].errorbar(df['jd'][(df['filter']=='ZTF_g')]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_g')],
             yerr = df['magunc'][(df['filter']=='ZTF_g')],
            marker='.',
            color='green')
axs[1].errorbar(df['jd'][(df['filter']=='ZTF_r')]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_r')],
             yerr = df['magunc'][(df['filter']=='ZTF_r')],
            marker='.',
            color='red')
axs[1].errorbar(df['jd'][(df['filter']=='ZTF_i')]-2400000.5, 
             df['mag'][(df['filter']=='ZTF_i')],
             yerr = df['magunc'][(df['filter']=='ZTF_i')],
            marker='.',
            color='purple')

peak = peaks[sn_name]
axs[0].set_xlim(peak-25, peak+75)
axs[1].set_xlim(peak-25, peak+75)

# axs[1].set_ylim(18, 20)
axs[0].set_xlabel('Time (MJD)')
axs[0].set_ylabel('Magnitude')
axs[0].set_title(sn_name+' SNR>3')
axs[0].axvline(peak)

axs[0].invert_yaxis()
axs[1].invert_yaxis()

axs[1].set_xlabel('Time (MJD)')
axs[1].set_ylabel('Magnitude')
axs[1].set_title(sn_name)
axs[1].axvline(peak)

# plt.savefig(path+sn_name+'.png')
```
