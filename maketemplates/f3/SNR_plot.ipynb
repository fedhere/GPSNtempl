```python
allsne = pd.read_csv(os.getenv("SESNPATH") +
                          "/utils/SESNessentials.csv", encoding = "ISO-8859-1")['SNname'].values
```

```python
'''
Here we are calculating SNR for each SN in each band as follows:
SNR = flux/dflux for each data point

We then collect SNRs of all the data points in bands as below:
['U', 'u', 'B', 'V', 'R', 'I', 'g', 'r', 'i', 'J', 'H', 'K', 'w1', 'w2', 'm2']

'''


temp = []

snrs = []
bands = []
names = []

fluxes = []

dmags_all = []
lcmags_all = []



lc_ = {}
# bands = su.bands

for i, sn in enumerate(allsne):
    
#     snr['sn'].append(sn)
    
    # read and set up SN and look for photometry files
    try:
        thissn = snstuff.mysn(sn, addlit=True)
    except AttributeError:
        print('removed sn: ', sn)
        continue
    if len(thissn.optfiles) + len(thissn.fnir) == 0:
        print ("bad sn")
    # read metadata for SN
    thissn.readinfofileall(verbose=False, earliest=False, loose=True)
    thissn.setVmax()

    lc, flux, dflux, snname = thissn.loadsn2()
    # set up photometry
    # thissn.printsn()
    thissn.setphot()
    thissn.getphot()
    thissn.setphase()
    
    # check SN is ok and load data
    if thissn.Vmax is None or thissn.Vmax == 0 or np.isnan(thissn.Vmax):
        print ("bad sn: No Vmax")
  

    # lc, flux, dflux, snname = thissn.loadsn2(verbose=True)
    
    
    
    
#     lc['dmag'][lc['dmag']==0] = 0.1*np.ones(len(lc['dmag'][lc['dmag']==0]))
    
    lc_mag = lc['ccmag']#[lc['dmag'] >= 0.01]
    lc_photcode = lc['photcode']#[lc['dmag'] >= 0.01]
    lc_dmag = lc['dmag']#[lc['dmag'] >= 0.01]
    
    print(lc['dmag'][lc['dmag'] < 0.01])
    
    
    
    flux = 10 ** (-lc_mag / 2.5) * 5e8
    dflux = flux * lc_dmag * LN10on2p5
    
        
    # flux = 10 ** (-lc_mag / 2.5) 
    # dflux = flux * np.log(10) * lc_dmag / 2.5
        
    
#     if len(lc['photcode']) != len(flux/dflux):
#         temp.append(sn)
    snr_ratio = flux/dflux
    
        
    
    photcode = [lc_photcode[i].decode().split('l')[0] for i in range(len(lc_photcode))]
    
    if len(np.setdiff1d(np.unique(photcode), su.bands, assume_unique=False)) != 0:
        # If there are bands that we don't want, remove their snrs.
        
        dd = (np.setdiff1d(np.unique(photcode), su.bands, assume_unique=False))
        for d in dd:
        
            snr_ratio = np.asarray(snr_ratio)[np.asarray(photcode) != str(d[0])]
            photcode = np.asarray(photcode)[np.asarray(photcode) != str(d[0])]
        
    

    
    thissn.setphot()
    thissn.getphot()
    thissn.setphase()
    thissn.sortlc()
    
    names_tmp = np.array([sn for i in range(len(snr_ratio))])
    
    
        
    snrs = np.concatenate((snrs, snr_ratio), axis = 0)
    fluxes = np.concatenate((fluxes, flux), axis = 0)
    bands = np.concatenate((bands, photcode), axis = 0)
    names = np.concatenate((names, names_tmp), axis = 0)
    dmags_all = np.concatenate((dmags_all, lc_dmag), axis = 0)
    lcmags_all = np.concatenate((lcmags_all, lc_mag), axis = 0)
    
plt.hist(np.log10(fluxes))
print(np.min(fluxes))
# plt.xscale('log')
    
```

```python
snr = pd.DataFrame({'SNR': snrs, 'Band': bands})
```
