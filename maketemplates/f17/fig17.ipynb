```python
colors = {"Ibc":"k", 
          "Ib":"SteelBlue", 
          "IIb":"FireBrick", 
          "Ic":"DarkGreen", 
          "Ic-bl":"DarkOrange", 
          "Ibn":"purple"
            }


```

```python
def plotDms(sntype, df, ax, label=None, offset=0.0, ylabel="", 
            grid=False, axis="on", rot=0, yoff=0, loc=None):
    
    
    for i in df.index:
        yerr = [df.loc[i, sntype] + df.loc[i, sntype+"-"] + yoff, 
                  df.loc[i, sntype] + df.loc[i, sntype+"+"] + yoff]
        ax.fill_betweenx(yerr, i*2 + offset - 0.05, i*2 + offset + 0.05, 
                         color = colors[sntype], alpha=0.5)
    ax.scatter((df.index)*2 + offset, df[sntype] + yoff, s = 50,
               edgecolor=colors[sntype], label=label, facecolor=colors[sntype])#"white")
    ax.plot((df.index)*2 + offset, df[sntype] + yoff, '--',
               color=colors[sntype])#"white")
            
    #ax.set_xticks(np.arange(len(df["filter"])*2)[::2] + 0.5, 
    #              df['filter'], rotation=rot) ;
    #plt.ylabel(r"$\Delta m_{15}$")
    if not label:
        ax.text(0.85, 0.85, sntype, transform=ax.transAxes)
    else:
        plt.legend(markerscale=2, fontsize=25)#, loc=loc)
    ax.grid(visible=grid)
```

```python
plt.rcParams['font.size'] = 35
fig4 = plt.figure(constrained_layout=True, figsize=(20,15))#, sharex=True, sharey=True)
spec4 = plt.GridSpec(ncols=1, nrows=4, hspace=0, wspace=0)
anno_opts = dict(xy=(0.5, 0.5), xycoords='axes fraction',
                 va='center', ha='center')

ax = fig4.add_subplot(spec4[0:-1, 0])
for i,tp in enumerate(["Ibc", "IIb", "Ib", "Ic", "Ic-bl", "Ibn"]):
    label = tp
    if i > 0: 
        label += " + %d"%i
    plotDms(tp, dm15, ax, label=label, offset=0, 
            ylabel=r"$\Delta m_{15}$", yoff=i, loc='upper left')

ylim = plt.ylim()
plt.yticks(range(*np.array(ylim).astype(int)))
ax.yaxis.grid(True)
plt.ylabel(r"$\Delta m_{15}$ + offset")


ax = fig4.add_subplot(spec4[3, 0])
offsets = []
for i,tp in enumerate(["Ibc", "IIb", "Ib", "Ic", "Ic-bl", "Ibn"]):
    label = ""
    ax.errorbar((dm15.index)*2, dm15[tp], yerr = [-dm15[tp + "-"], 
                                                  dm15[tp + "+"]],
                color=colors[tp],
            fmt=' ', label=label, lw=4, alpha=0.5, ls='none', ms=0)
    ax.scatter((dm15.index)*2, dm15[tp], lw=2, 
               edgecolor=colors[tp], facecolor="white")
    
ax.set_xticks(np.arange(len(dm15["filter"])*2)[::2], 
                  dm15['filter']) ;
ax.set_ylim(-0.6,2.9)
ax.set_yticks([0,1,2])
ax.yaxis.grid(True)
plt.ylabel(r"$\Delta m_{15}$")

plt.savefig("dm15.png")
```

```python
def plotDms_leff(sntype, df, ax, label=None, offset=0.0, ylabel="", 
            grid=False, axis="on", rot=0, yoff=0, loc=None):
    
    
    for i in df.index: # for every band
        yerr = [df.loc[i, sntype] + df.loc[i, sntype+"-"] + yoff, 
                  df.loc[i, sntype] + df.loc[i, sntype+"+"] + yoff]
        l = leff[df.loc[i, "filter"]]
        ax.fill_betweenx(yerr, l + offset - 50, l + offset + 50, 
                         color = colors[sntype], alpha=0.5)
    ax.scatter([leff[f] for f in df["filter"]], df[sntype] + yoff, s = 50,
               edgecolor=colors[sntype], label=label, facecolor=colors[sntype])#"white")
    ax.plot([leff[f] for f in df["filter"]], df[sntype] + yoff, '-',
               color=colors[sntype])#"white")
            
    #ax.set_xticks(np.arange(len(df["filter"])*2)[::2] + 0.5, 
    #              df['filter'], rotation=rot) ;
    #plt.ylabel(r"$\Delta m_{15}$")
    if not label:
        ax.text(0.85, 0.85, sntype, transform=ax.transAxes)
    else:
        plt.legend(markerscale=2, fontsize=25)#, loc=loc)
    ax.grid(visible=grid)
```
